# ①ライブラリのインポート

```python
# データハンドリング
import numpy as np
import pandas as pd
# 正規化
from sklearn.preprocessing import MinMaxScaler
# データ分割
from sklearn.model_selection import train_test_split
# ニューラルネット
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
# 混同行列可視化用
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
# 正解率の算出
from sklearn.metrics import accuracy_score
# 決定木可視化用
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image
```

# ②データの読み込みと分割

```python
# 読み込み
data=pd.read_csv('data_complete.csv')

#説明変数と目的変数に分割
X=data[
    [
        'Pclass', 'Age', 'SibSp', 'Parch', 
        'Fare', 'Sex_female','Sex_male', 
        'Embarked_C', 'Embarked_Q', 'Embarked_S'
    ]
]
y=data['Survived']

# 訓練データとテストデータに分割
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1234)

# 正規化
scaler_x = MinMaxScaler()
X_train = scaler_x.fit_transform(X_train)
X_test = scaler_x.transform(X_test) 
```

# ③決定木

1.決定木を用いて、目的変数の値に最も影響を与える変数を調査する。

2.影響が強い変数が判明すれば、その変数を基にデータを分割する。

3.分割したデータでそれぞれモデルを最適化してあげる


という手法は、一般的に用いられるが、それはデータ数が豊富な時に限られる。

今回のようなデータが少ない問題では、学習データを減らし、十分な学習を行えなくする可能性があるので、1つの手法として「こんなのがあるんだー」くらいに思っておいてほしい。

### 決定木の使い道

データ分割ではなく、原因推定や施策決定に用いられるのが一般的です。

#### 例：ゲームのログ解析

あるゲームアプリのユーザの離脱率を様々な変数を用いて、決定木にかけると、「インストール翌日にログインしたか否か」という変数が上位変数に上がってくれば、インストール翌日にログインしてもらうような魅力的な施策を提案する必要があるという風なことがわかる。

```python
"""基本モデルの作成
"""
from sklearn import tree
clf=tree.DecisionTreeClassifier(max_depth=3)
```

```python
"""モデルの最適化
"""
clf.fit(X_train,y_train)
```

```python
"""データの可視化
"""
# 可視化用オブジェクト作成
dot_data=tree.export_graphviz(
    clf,
    out_file=None,                      #ファイル名
    filled=True,                       #色を塗る
    feature_names=X.columns,           #特徴量(説明変数の名前)
    class_names=['survived', 'death'] #目的変数の分類名
)
# 可視化情報生成
graph=pydotplus.graph_from_dot_data(dot_data)
# 可視化実行
Image(graph.create_png())
```

# ⑤性別でモデルを分割してみる

```python
data.columns
```

```python
"""データの準備
"""
# male/femaleを分割
data_male=data.loc[
    data['Sex_male']==1,
    [
        'Survived','Pclass', 'Age', 'SibSp', 
        'Parch', 'Fare','Sex_male', 
        'Embarked_C', 'Embarked_Q', 'Embarked_S'
    ]
]
data_female=data.loc[
    data['Sex_female']==1,
    [
        'Survived','Pclass', 'Age', 'SibSp', 
        'Parch', 'Fare', 'Sex_female', 
        'Embarked_C', 'Embarked_Q', 'Embarked_S'
    ]
]

# male/femaleのそれぞれの説明変数と目的変数に分割
X_male=data_male[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_C', 'Embarked_Q', 'Embarked_S']]
X_female=data_female[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_female', 'Embarked_C', 'Embarked_Q', 'Embarked_S']]
y_male=data_male['Survived']
y_female=data_female['Survived']

# male/femaleのそれぞれの訓練データとテストデータに分割
X_train_m,X_test_m,y_train_m,y_test_m=train_test_split(X_male,y_male,test_size=0.2,random_state=1234)
X_train_f,X_test_f,y_train_f,y_test_f=train_test_split(X_female,y_female,test_size=0.2,random_state=1234)

# 正規化
scaler_x = MinMaxScaler()
X_train_m = scaler_x.fit_transform(X_train_m)
X_test_m = scaler_x.transform(X_test_m) 
X_train_f = scaler_x.fit_transform(X_train_f)
X_test_f = scaler_x.transform(X_test_f) 
```

```python
"""基本モデルの作成
"""
clf_m=MLPClassifier(
    random_state=1234
)
clf_f=MLPClassifier(
    random_state=1234
)
```

```python
"""モデルの最適化
"""
clf_m.fit(X_train_m,y_train_m)
clf_f.fit(X_train_f,y_train_f)
```

```python
"""予測値算出
"""
y_train_predict_m=clf_m.predict(X_train_m)
y_train_predict_f=clf_f.predict(X_train_f)
```

```python
"""精度の評価
"""
y_train=np.concatenate([y_train_m,y_train_f])
y_train_predict=np.concatenate([y_train_predict_m,y_train_predict_f])
print("正解率:",round(accuracy_score(y_train, y_train_predict),3))
```

# ⑥テストデータへの適用

```python
"""精度の評価
"""
# 予測値算出
y_test_predict_m=clf_m.predict(X_test_m)
y_test_predict_f=clf_f.predict(X_test_f)
# 精度算出
y_test=np.concatenate([y_test_m,y_test_f])
y_test_predict=np.concatenate([y_test_predict_m,y_test_predict_f])
print("正解率:",round(accuracy_score(y_test, y_test_predict),3))
```
