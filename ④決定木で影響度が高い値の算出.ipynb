```python
"""最も生存率の高いEmbarked(出港地) = S で補完
"""
data=data.fillna({"Embarked":"S"})
```

### ●ダミー変数化

```python
"""Sex(性別)の置換
"""
# male=0   female=1  に置換
data["Sex"][data["Sex"]=='male']=0
data["Sex"][data["Sex"]=='female']=1
```

```python
"""Embarked(出港地)の置換
"""
data["Embarked"]=pd.get_dummies(data["Embarked"])
```

### ●データの分割

```python
#説明変数(Pclass,Sex,Age,SibSp,Parch,Fare,Embarked)
X=data.iloc[0:,[2,4,5,6,7,9,11]]
#目的変数(Survived)
y=data.iloc[0:,[1]]

"""訓練データとテストデータに分割(7:3)
"""
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1234)
```

### ●決定木

```python
"""基本モデルの作成
"""
from sklearn import tree
clf=tree.DecisionTreeClassifier(max_depth=3)
```

```python
"""y_train,y_testを一次元に変換
"""
y_train=y_train.values.flatten()
y_test=y_test.values.flatten()
```

```python
"""モデルの最適化
"""
#説明変数:X_train    目的変数:y_train
clf.fit(X_train,y_train)
```

```python
"""graphviz,pydotplusのインポート
"""
from sklearn.tree import export_graphviz
import pydotplus
```

```python
"""データの可視化
"""
dot_data=tree.export_graphviz(
    clf,
    out_file=None, #ファイル名
    filled=True, #色を塗る
    feature_names=X.columns, #特徴量(説明変数の名前)
    class_names=['survived', 'death'] #目的変数の分類名
)

graph=pydotplus.graph_from_dot_data(dot_data)
```

```python
"""表示
"""
from IPython.display import Image
Image(graph.create_png())
```
